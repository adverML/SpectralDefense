```python
import matplotlib.pyplot as plt 
```

```python
eps = [0.001,0.003, 0.01,0.03, 0.1, 0.3,1]
attack_fgsm = [0.100648,0.31299543,  0.747057, 0.95928,0.9156858 ,0.9016334,0.898390]
detect_fgsm = [0.5052437641723356,0.54651282,0.86082, 0.99752,1.0,1.0, 1.0]
attack_bim = [0.105092,0.36043, 0.91712, 1.0,1.0 , 1.0 ,1.0]
detect_bim = [0.50426122,0.5443894,0.7382798, 0.978047,0.999907,1.0, 1.0]
attack_pgd = [0.09031948,0.3105933,  0.862719, 1.0,1.0 , 1.0 ,1.0]
detect_pgd = [0.502697,0.5291177,0.7472809, 0.9785145,0.99991,1.0, 1.0]
attack_aa  = [0.1115,  0.3792,  0.9479,  0.99    , 1.0 , 1.0    ,1.0]
detect_aa =  [0.5053,  0.5397,0.7491, 0.9998,0.9998 ,1.0, 1.0]
```

```python
plt.figure(figsize=(10,5))
plt.title('Success rates for attacks and Black-Box detection',fontsize=20)
plt.semilogx(eps, attack_fgsm, label='FGSM Attack rate', linestyle='--', color='b')
plt.semilogx(eps, detect_fgsm, label='FGSM Detect rate', linestyle='-', color='b')
plt.semilogx(eps, attack_bim,  label='BIM Attack rate', linestyle='--', color='r')
plt.semilogx(eps, detect_bim,  label='BIM Detect rate', color='r')
plt.semilogx(eps, attack_pgd,  label='PGD Attack rate', linestyle='--', color='g')
plt.semilogx(eps, detect_pgd,  label='PGD Detect rate', color='g')
plt.semilogx(eps, attack_aa ,  label='AA  Attack rate', linestyle='--', color='cyan')
plt.semilogx(eps, detect_aa ,  label='AA  Detect rate', color='cyan')

plt.xlabel('Epsilon $\epsilon$',    fontsize=14)
plt.ylabel('Success Rate (in AUC)', fontsize=14)

plt.margins(tight=True)
plt.legend(fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

plt.savefig('epsilons.png', bbox_inches='tight')
```
