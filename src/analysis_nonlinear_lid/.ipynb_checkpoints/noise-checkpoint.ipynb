```python
def gen_noisy_image(image, mean=0, var=4./255.):
    row,col,ch= image.shape
    mean = 0
    var = 4./255. 
    sigma =  var #**0.5
    # noisy = gaussian_filter(image, sigma)
    gauss = np.random.normal(mean,sigma, size=(row,col,ch)).astype('float32')
    noisy = np.clip(image + gauss, 0, 1) # as FGSM
    
    return noisy
```
