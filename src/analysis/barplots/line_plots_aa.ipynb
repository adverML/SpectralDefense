```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.figure(10)

# create data
df_bb = pd.DataFrame({
    'Gauss': [28.75,    22.5,   22.33], 
     'FGSM': [ 0.33,    0.0,     0.0 ], 
     'BIM' : [12.67,    3.33,    1.33], 
     'PGD' : [11.67,    4.33,    2.67],
     'AA'  : [11.00,    4.33,    0.67], 
     'DF'  : [54.17,   48.33,   51.67],
     'CW'  : [41.67,   46.00,   41.33],
      }, index=["32", "64", "128"]
    )

df_wb = pd.DataFrame({
    'Gauss': [0.26,   0.0,   0.0 ], 
    'FGSM':  [0.0,    0.0,   0.0 ], 
    'BIM' :  [0.0,   0.0,   0.33 ], 
    'PGD' :  [0.0,   0.0,   0.0 ],
    'AA'  :  [0.0,   0.0,   0.33 ], 
    'DF'  :  [53.33,  54.67, 53.33 ],
    'CW'  :  [47.67,  51.33, 55.67 ],
      }, index=["32", "64", "128"]
    )

color = ['pink', 'blue', 'red','green','cyan','orange', 'purple']

# view data
print(df_bb)

ax = plt.gca()

# plot grouped bar chart
ax1 = df_bb.plot.line(
            ax=ax,
            # x='DataSet',
            xlabel="ImageNet - Image Size",
            # rot=0,
            legend=False,
            color=color,
            alpha=0.8,
            marker='v',
            title="ImageNet - Random Forest - BB/WB", 
            ylabel='ASRD',
            ylim=[0, 58.0]
        )

ax2 = df_wb.plot.line(
    ax=ax, 
    legend=False,
    marker='o', 
    linestyle='--', 
    color=color
    )

# legend = ax2.legend( title='WB',  loc='center right', framealpha=0.3)
legen2 = plt.legend(df_bb, title='BB', loc='center', framealpha=0.3)
plt.gca().add_artist(legen2)

h, l = ax2.get_legend_handles_labels()
legend = plt.legend(h[7:14], l[7:14], title='WB',  loc='center right', framealpha=0.3)

# ax.set_xticks([32, 64 , 128])
ax.grid(which='major', alpha=0.2)
# plt.show()



plt.savefig("line_im.png", bbox_inches='tight')

```
